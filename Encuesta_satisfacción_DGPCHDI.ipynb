```python
#modelo_final
import pandas as pd
import random
from math import ceil

# Step 1: Load the Excel file with the list of companies
file_path = '/content/N_2024_input.xlsx'
df = pd.read_excel(file_path)

# Step 2: Load the Excel file with the list of emails
file_path_domicilios = '/content/ListadoDomicilios.xlsx'
df_domicilios = pd.read_excel(file_path_domicilios)

# Step 3: Extract the list of companies from the "Empresa_agosto" column
empresas = df['Empresa_setiembre'].tolist()

# Step 4: Define the size of the universe (N)
N = len(empresas)

# Step 5: Set parameters for sample size calculation
Z = 1.28  # Corresponds to a 80% confidence level
E = 0.15  # 15% margin of error
p = 0.5   # Estimated proportion of success

# Step 6: Calculate the sample size (n) using the provided formula
numerator = N * (Z**2) * p * (1 - p)
denominator = (E**2) * (N - 1) + (Z**2) * p * (1 - p)
n = ceil(numerator / denominator)

# Step 7: Select companies randomly for the sample
muestra = random.sample(empresas, n)


# Step 8: Add a new column 'seleccion_julio' to the DataFrame
df['seleccion_setiembre'] = df['Empresa_setiembre'].apply(lambda x: 'seleccionada' if x in muestra else '')

# Step 9: First match based on RUC2
email_dict = df_domicilios.set_index('RUC2')['direccion_electronica'].to_dict()
df['correo_setiembre'] = df['Empresa_setiembre'].map(email_dict)

# Step 10: Second match based on NOMBRE_COMPLETO1 (if no match found in first step)
def invert_name_order(name):
    if isinstance(name, str):
        parts = name.split()
        if len(parts) == 4:
            # The pattern is assumed to be "First Name Second Name Last Name1 Last Name2"
            return f"{parts[2]} {parts[3]} {parts[0]} {parts[1]}"
    return name  # Return the original name if the pattern is not matched

df_domicilios['NOMBRE_COMPLETO1_INVERTED'] = df_domicilios['NOMBRE_COMPLETO1'].apply(invert_name_order)
name_email_dict = df_domicilios.set_index('NOMBRE_COMPLETO1_INVERTED')['direccion_electronica'].to_dict()

# Apply the second match where the first match failed
df['correo_setiembre'] = df.apply(lambda row: name_email_dict.get(row['Empresa_setiembre'], row['correo_setiembre']) if pd.isna(row['correo_setiembre']) else row['correo_setiembre'], axis=1)

# Step 11: Save the updated DataFrame back to an Excel file
output_file_path = '/content/N_2024_output.xlsx'
df.to_excel(output_file_path, index=False)

print(f"Archivo guardado como {output_file_path}")

```
