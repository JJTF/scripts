```python
import tabula as tb
import pandas as pd

file = "PDF YAMASAN (2)_rotated.pdf"

# Extraer datos desde la página 1 hasta la página 2
#tables = tb.read_pdf(file, pages="1-12", multiple_tables=True)

tables = tb.read_pdf(file, pages="1-3", multiple_tables=True)

# Convertir las tablas extraídas en un solo DataFrame
df_combined = pd.concat(tables, ignore_index=True)

# Guardar el DataFrame en un archivo Excel
df_combined.to_excel("tabla_1.xlsx", index=False)
```
