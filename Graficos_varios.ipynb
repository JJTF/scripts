```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos
file_path = '/content/bd_poi_ene_may_2024.xlsx'
data = pd.read_excel(file_path)

# Meses
months = ["ENERO", "FEBRERO", "MARZO", "ABRIL", "MAYO", "JUNIO",
          "JULIO", "AGOSTO", "SETIEMBRE", "OCTUBRE", "NOVIEMBRE", "DICIEMBRE"]

# Filtrar los datos para las metas 87 y 95
meta_87_data = data[data['META'] == 87]
meta_95_data = data[data['META'] == 95]

# Sumar los valores mensuales para cada meta
monthly_values_meta_87 = meta_87_data[months].sum()
monthly_values_meta_95 = meta_95_data[months].sum()

# Calcular el acumulado mensual para cada meta
cumulative_meta_87 = monthly_values_meta_87.cumsum()
cumulative_meta_95 = monthly_values_meta_95.cumsum()

# Calcular el porcentaje del PIM total para cada acumulado mensual
pim_value_meta_87 = meta_87_data['PIM'].sum()
pim_value_meta_95 = meta_95_data['PIM'].sum()
percentage_cumulative_meta_87 = (cumulative_meta_87 / pim_value_meta_87) * 100
percentage_cumulative_meta_95 = (cumulative_meta_95 / pim_value_meta_95) * 100

# Crear el gr√°fico de barras apiladas con acumulados mensuales
fig, ax = plt.subplots(figsize=(14, 8))

# Plot the cumulative values for Meta 87 and Meta 95
bar_width = 0.35
index = range(len(months))

bars_meta_87 = ax.bar(index, cumulative_meta_87, bar_width, label='Meta 87')
bars_meta_95 = ax.bar([i + bar_width for i in index], cumulative_meta_95, bar_width, label='Meta 95')

# Adding title and labels
ax.set_title('Acumulado Mensual por Meta 2024')
ax.set_xlabel('Meses')
ax.set_ylabel('Monto Acumulado en S/.')
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(months, rotation=90)
ax.legend()

# Adding text labels on top of the bars for Meta 87
for bar, value, percentage in zip(bars_meta_87, cumulative_meta_87, percentage_cumulative_meta_87):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height,
            f'S/. {value:.2f}\n{percentage:.1f}%',
            ha='center', va='bottom', fontsize=10, rotation=90)

# Adding text labels on top of the bars for Meta 95
for bar, value, percentage in zip(bars_meta_95, cumulative_meta_95, percentage_cumulative_meta_95):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height,
            f'S/. {value:.2f}\n{percentage:.1f}%',
            ha='center', va='bottom', fontsize=10, rotation=90)

# Mostrar el gr√°fico
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo Excel
file_path = '/content/bd_poi_ene_may_2024.xlsx'
data = pd.read_excel(file_path)

# Filtrar los datos para la meta 95
meta_95_data = data[data['META'] == 95]

# Sumar los valores de PIM para cada CLASIFICADOR1 dentro de la GENERICA "3.BIENES Y SERVICIOS" para la META 95
pim_meta_95_bienes_clasificador = meta_95_data[meta_95_data['GENERICA'] == '3.BIENES Y SERVICIOS'].groupby('CLASIFICADOR1')['PIM'].sum()

# Calcular el porcentaje de cada CLASIFICADOR1 dentro de la GENERICA "3.BIENES Y SERVICIOS" para la META 95
total_pim_meta_95_bienes = pim_meta_95_bienes_clasificador.sum()
percent_meta_95_bienes_clasificador = (pim_meta_95_bienes_clasificador / total_pim_meta_95_bienes) * 100

# Limpiar los nombres de los clasificadores (eliminando los n√∫meros que anteceden al clasificador)
cleaned_labels = pim_meta_95_bienes_clasificador.index.str.replace(r'^\d+\.', '', regex=True)

# Crear gr√°fico de dona para desglosar "Bienes y Servicios" en la META 95
labels_meta_95_bienes = [f'{clas} (S/. {pim:.2f}) - {perc:.1f}%' for clas, pim, perc in zip(cleaned_labels, pim_meta_95_bienes_clasificador, percent_meta_95_bienes_clasificador)]
sizes_meta_95_bienes = percent_meta_95_bienes_clasificador

fig3, ax3 = plt.subplots()
ax3.pie(sizes_meta_95_bienes, labels=labels_meta_95_bienes, autopct='%1.1f%%', startangle=90, pctdistance=0.85, explode=[0.05]*len(sizes_meta_95_bienes))

# A√±adir un c√≠rculo blanco en el centro para crear un gr√°fico de dona
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig3.gca().add_artist(centre_circle)

# Asegurar que el gr√°fico de dona sea circular
ax3.axis('equal')

plt.title('Distribuci√≥n del PIM por Clasificador1 - Meta 95 (Bienes y Servicios)')
plt.show()
```
