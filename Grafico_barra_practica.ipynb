```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Datos de muestra
categorias = ['Lima Centro', 'Lima Este', 'Lima Norte', 'Lambayeque', 'Arequipa']
cantidades = [440164, 333703, 265879, 245209, 243452]
porcentajes=[10.9, 8.26, 6.58,6.07,6.03]
```

```python
"total = sum(cantidades)
"porcentajes = [(cantidad / total) * 100 for cantidad in cantidades]
```

```python
# Configuración del gráfico de barras
fig, ax1 = plt.subplots()
ax1.bar(categorias, cantidades, label='Cantidad')
ax1.set_xlabel('Categorías')
ax1.set_ylabel('Cantidad')
# Configuración adicional del gráfico de barras
ax1.set_title('Gráfico de barras y línea de porcentajes')
#ax1.legend(loc='upper left')
ax1.set_xticklabels(categorias, rotation=90)
# Crea el gráfico de línea para los porcentajes
ax2 = ax1.twinx()
ax2.plot(categorias, porcentajes, color='red', marker='o', label='Porcentaje')
ax2.set_ylabel('Porcentaje')
ax2.legend(loc='upper right')
# Agrega las etiquetas de cantidad en el gráfico de barras
for i, cantidad in enumerate(cantidades):
    ax1.text(i, cantidad + 1, str(cantidad), ha='center')
# Agrega las etiquetas de porcentaje en el gráfico de línea
for i, porcentaje in enumerate(porcentajes):
    ax2.text(i, porcentaje + 1, f'{porcentaje:.2f}%', ha='center')
#
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper right')
# Muestra el gráfico
plt.show()
```
