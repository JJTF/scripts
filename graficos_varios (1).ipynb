```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo Excel
data = pd.read_excel("prueba_1.xlsx")  # Reemplaza con el nombre de tu archivo

# Obtener los datos de las columnas necesarias
years = data['var']
ingresados = data['ingresado']
atendidos = data['atendido']

# Calcular los porcentajes
porcentajes = [(atendidos[i] / ingresados[i]) * 100 for i in range(len(years))]

# Crear la figura y los ejes
fig, ax1 = plt.subplots()

# Graficar las barras con etiquetas en forma de cuadro
bars_1 = ax1.bar(years, ingresados, color='blue', label='Ingresados')
bars_2 = ax1.bar(years, atendidos, color='orange', label='Atendidos')

for bar_1, ingresado in zip(bars_1, ingresados):
    yval = bar_1.get_height()
    ax1.text(bar_1.get_x() + bar_1.get_width()/2, yval, int(ingresado),
             ha='center', va='bottom', color='black', bbox=dict(facecolor='white', alpha=0.7))

for bar_2, atendido in zip(bars_2, atendidos):
    yval = bar_2.get_height()
    ax1.text(bar_2.get_x() + bar_2.get_width()/2, yval, int(atendido),
             ha='center', va='bottom', color='black', bbox=dict(facecolor='white', alpha=0.7))

# Configurar el eje y1 (izquierdo) para las barras
ax1.set_xlabel('Año')
ax1.set_ylabel('Cantidad')
ax1.set_title('')
ax1.legend(loc='upper left')

# Crear el eje y2 (derecho) para los porcentajes
ax2 = ax1.twinx()
line = ax2.plot(years, porcentajes, color='red', marker='o', label='Porcentaje')[0]
for i, txt in enumerate(porcentajes):
    ax2.annotate(f"{txt:.0f}%", (years[i], porcentajes[i]), textcoords="offset points", xytext=(0,20), ha='center', fontsize=8, color='black')

ax2.set_ylabel('Porcentaje (%)')
ax2.legend(loc='upper right')

# Mover la leyenda del porcentaje fuera del gráfico
ax2.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Mostrar el gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo Excel
data = pd.read_excel("prueba_1.xlsx")  # Reemplaza con el nombre de tu archivo

# Obtener los datos de las columnas necesarias
years = data['var']
var_1 = data['ingresado']
var_2 = data['atendido']

# Calcular la variable calculada (var_2 / var_1) en porcentajes
var_calculada = (var_2 / var_1) * 100

# Crear la figura y los ejes
fig, (ax1, ax2) = plt.subplots(nrows=2, sharex=True, figsize=(10, 8))

# Graficar var_1 y var_2 en el primer eje (ax1) con etiquetas
ax1.plot(years, var_1, label='Ingresado', marker='o')
ax1.plot(years, var_2, label='Atendido', marker='s')
ax1.set_ylabel('Valor')
ax1.set_title('Ingresado y Atendido por año')
ax1.legend(loc='upper left')
for i, txt in enumerate(var_1):
     ax1.annotate(f"{int(txt):,}", (years[i], var_1[i]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=8, color='black')
for i, txt in enumerate(var_2):
     ax1.annotate(f"{int(txt):,}", (years[i], var_2[i]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=8, color='black')

# Graficar la variable calculada en el segundo eje (ax2) con etiquetas
ax2.plot(years, var_calculada, label='% Atendidos', color='red', marker='x')
ax2.set_xlabel('Año')
ax2.set_ylabel('Porcentaje (%)')
ax2.legend(loc='upper left')
for i, txt in enumerate(var_calculada):
    ax2.annotate(f"{txt:.2f}%", (years[i], var_calculada[i]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=8, color='black')

# Mover la leyenda fuera de los gráficos
ax1.legend(loc='upper left', bbox_to_anchor=(1, 1))
ax2.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Ajustar el diseño para evitar recorte de la leyenda
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo Excel
data = pd.read_excel("prueba_1.xlsx")  # Reemplaza con el nombre de tu archivo

# Obtener los datos de las columnas necesarias
years = data['var']
var_1 = data['ingresado']
var_2 = data['atendido']

# Calcular la variable calculada (var_2 / var_1) en porcentajes
var_calculada = (var_2 / var_1) * 100

# Crear la figura y el eje
fig, ax = plt.subplots(figsize=(10, 6))

# Graficar var_1 y var_2 con etiquetas
ax.plot(years, var_1, label='Ingresado', marker='o')
ax.plot(years, var_2, label='Atendido', marker='s')
ax.set_ylabel('Valor')

# Graficar la variable calculada con etiquetas en porcentajes
ax2 = ax.twinx()
ax2.plot(years, var_calculada, label='Atendido (%)', color='red', marker='x')
ax2.set_ylabel('Porcentaje (%)')

# Agregar etiquetas a los puntos de datos con desplazamiento vertical
for i, txt in enumerate(var_1):
    ax.annotate(f"{int(txt):,}", (years[i], var_1[i]), textcoords="offset points", xytext=(0,10), ha='left', fontsize=8, color='black')
for i, txt in enumerate(var_2):
    ax.annotate(f"{int(txt):,}", (years[i], var_2[i]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=8, color='black')
    if i > 0 and abs(var_2[i] - var_2[i-1]) < 500:  # Aplicar desplazamiento si la diferencia es pequeña
        ax.annotate(f"{int(txt):,}", (years[i], var_2[i]), textcoords="offset points", xytext=(0,15), ha='center', fontsize=8, color='black')
for i, txt in enumerate(var_calculada):
    ax2.annotate(f"{txt:.2f}%", (years[i], var_calculada[i]), textcoords="offset points", xytext=(0,1), ha='right', fontsize=8, color='black')

# Ajustar el diseño y las leyendas
lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left', bbox_to_anchor=(1, 1))
ax.set_xlabel('Año')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Cargar los datos desde el archivo Excel
data = pd.read_excel("prueba_1.xlsx")  # Reemplaza con el nombre de tu archivo

# Obtener los datos de las columnas necesarias
years = data['var']
var_1 = data['ingresado']
var_2 = data['atendido']

# Calcular la variable calculada (var_2 / var_1) en porcentajes
var_calculada = (var_2 / var_1) * 100

# Crear la figura y el eje
fig, ax = plt.subplots(figsize=(10, 6))

# Crear un arreglo para las posiciones de las barras agrupadas
bar_width = 0.4
positions = np.arange(len(years))

# Graficar las variables como barras agrupadas con etiquetas en números enteros
bars1 = ax.barh(positions - bar_width/2, var_1, bar_width, label='Ingresado')
bars2 = ax.barh(positions + bar_width/2, var_2, bar_width, label='Atendido')
ax.set_yticks(positions)
ax.set_yticklabels(years)
ax.set_xlabel('Valor')
ax.set_ylabel('Año')

# Agregar etiquetas a las barras con números enteros
for bar in bars1:
    xval = bar.get_width()
    ax.text(xval, bar.get_y() + bar.get_height()/2, f"{int(xval):,}",
             va='center', ha='left', fontsize=8, color='black')
for bar in bars2:
    xval = bar.get_width()
    ax.text(xval, bar.get_y() + bar.get_height()/2, f"{int(xval):,}",
             va='center', ha='left', fontsize=8, color='black')

# Agregar la variable calculada como una barra adicional
ax2 = ax.twiny()
ax2.barh(positions, var_calculada, bar_width, color='red', label='Atendido (%)')
ax2.set_xlabel('Porcentaje (%)')

# Ajustar el diseño y las leyendas
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
ax2.legend(loc='upper left', bbox_to_anchor=(1, 0.9))

# Mostrar el gráfico
plt.title('xxxx')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import Normalize

# Cargar los datos desde el archivo Excel
data = pd.read_excel("prueba_1.xlsx")  # Reemplaza con el nombre de tu archivo

# Obtener los datos de las columnas necesarias
years = data['var']
var_1 = data['ingresado']
var_2 = data['atendido']

# Calcular la variable calculada (var_2 / var_1) en porcentajes
var_calculada = (var_2 / var_1) * 100

# Calcular la diferencia entre Variable_1 y Variable_2
difference = var_1 - var_2

# Crear la figura y el eje
fig, ax = plt.subplots(figsize=(10, 6))

# Graficar el gráfico de dispersión
sc = ax.scatter(var_1, var_2, c=difference, s=var_calculada, cmap='coolwarm', alpha=0.7, edgecolors='k')
ax.set_xlabel('Ingresado')
ax.set_ylabel('Atendido')
ax.set_title('Gráfico de Dispersión')

# Agregar etiquetas a los puntos de datos con números enteros y porcentajes
for i, txt in enumerate(years):
    ax.annotate(f"Año {int(txt)}\n{var_calculada[i]:.2f}%", (var_1[i], var_2[i]), textcoords="offset points", xytext=(5,5), fontsize=8, color='black')

# Añadir una barra de color
cbar = plt.colorbar(sc, ax=ax)
cbar.set_label('Diferencia (Ingresados - Atendidos)')

# Ajustar el diseño
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
